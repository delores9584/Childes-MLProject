```python
from sklearn.model_selection import train_test_split
from keras.preprocessing.text import Tokenizer, text_to_word_sequence
from keras.preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
from keras.layers import Embedding, Flatten, Dense
from keras.models import Sequential

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import keras
from keras import models, layers, regularizers
```

```python
data = pd.read_csv("Gillam_fin.csv")
data = data.drop("Unnamed: 0", axis = 1)
pd.DataFrame.head(data)
len(data.dropna()["gloss"])
```

```python
# Split train, val, test
trainval, test = train_test_split(data.dropna(), test_size=0.15, random_state=1234, shuffle = True)
train, val = train_test_split(trainval, test_size = 0.15, random_state = 1234, shuffle = True)

# Process POS tags as X variable
str_train = list(train["gloss"])
str_val = list(val["gloss"])
str_test = list(test["gloss"])

Imp_train = train["Impaired"]
Imp_val = val["Impaired"]
Imp_test = test["Impaired"]
```

```python
# Build vocabulary
tokenizer = Tokenizer()
tokenizer.fit_on_texts(list(data.dropna()["gloss"]))

# Transform list of strings to list of word sequences
train_seq = []
for lst in str_train:
    seq = text_to_word_sequence(lst)
    train_seq.append(seq)
word_idx = tokenizer.word_index
```

## No Pre-trained Embedding

```python
# Tokenize word strings and pad to length 50
max_len = 50

def to_tokens(text_data):
    to_seq = tokenizer.texts_to_sequences(text_data)
    x_data = pad_sequences(to_seq, maxlen = max_len)
    return x_data

X_train = to_tokens(str_train)
X_val = to_tokens(str_val)
X_test = to_tokens(str_test)
```

## Glove

```python
import os

# parsing the Glove word-embeddings (from the book)
glove_dir = './'

embeddings_index = {}
f = open(os.path.join(glove_dir, 'glove.6B.50d.txt'))
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()
print('Found {} word vectors.'.format(len(embeddings_index)))

# preparing Glove word-embeddings matrix (from the book)
max_words = len(word_idx)
embedding_dim = 50

EmbedMat_glove = np.zeros((max_words, embedding_dim))
for word, i in word_idx.items():
    if i < max_words:
        embedding_vector = embeddings_index.get(word)
        if embedding_vector is not None:
            EmbedMat_glove[i] = embedding_vector
print(max_words, embedding_dim)
```

## Word2Vec

```python
import gensim
from gensim.models import word2vec

# CBOW-based Word2Vec
mod_cbow = word2vec.Word2Vec(train_seq, size = embedding_dim, sg = 0)

## Take all unknown words to have word vectors of the average of the infrequent words
inf_words_cbow = mod_cbow.wv.index2entity[-embedding_dim:]
UNK_vec_cbow = np.average(mod_cbow.wv[inf_words_cbow], axis=0)

EmbedMat_cbow = np.zeros((max_words, embedding_dim))
for word, i in word_idx.items():
    if i < max_words:
        if word in mod_cbow.wv.vocab:
            embed_vec = mod_cbow.wv[word]
            EmbedMat_cbow[i] = embed_vec
        else:
            EmbedMat_cbow[i] = UNK_vec_cbow
            
# SG-based Word2Vec
mod_sg = word2vec.Word2Vec(train_seq, size = embedding_dim, sg = 1)
mod_sg_wv = mod_sg.wv

## Take all unknown words to have word vectors of the average of the infrequent words
inf_words_sg = mod_sg.wv.index2entity[-embedding_dim:]
UNK_vec_sg = np.average(mod_sg.wv[inf_words_sg], axis=0)

EmbedMat_sg = np.zeros((max_words, embedding_dim))
for word, i in word_idx.items():
    if i < max_words:
        if word in mod_sg.wv.vocab:
            embed_vec = mod_sg.wv[word]
            EmbedMat_cbow[i] = embed_vec
        else:
            EmbedMat_cbow[i] = UNK_vec_sg
```

## Input: word embeddings, Output: Impairment
### Baseline Naive: 3 hidden layers, 32 nodes

```python
num_norm = data["Impaired"].value_counts()[0]
num_imp = data["Impaired"].value_counts()[1]
total = num_norm+num_imp
NtoI = float(num_norm)/float(num_imp)
NtoI
```

```python
# Baseline model
from keras import optimizers, regularizers

naive = Sequential()
naive.add(Embedding(max_words, 32, input_length=50))
naive.add(Flatten())
naive.add(Dense(32, activation='relu'))
naive.add(Dense(32, activation='relu'))
naive.add(Dense(32, activation='relu'))
naive.add(Dense(1, activation='sigmoid'))
naive.summary()

callback_lst = [
    keras.callbacks.EarlyStopping(
            monitor = 'acc',
            patience = 3),
    keras.callbacks.ModelCheckpoint(
            filepath="best_model.h5",
            monitor = "val_loss",
            save_best_only = True)]

opt = optimizers.SGD(lr=0.001)
naive.compile(optimizer=opt,
              loss='binary_crossentropy',
              metrics=['acc'])
naive_history = naive.fit(X_train, Imp_train, epochs=50, batch_size=512,
                          callbacks = callback_lst, shuffle = True,
                          validation_data=(X_val, Imp_val), class_weight={0: 3.380773819669337, 1: 1})
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['axes.titlepad'] = 20

naive_dict = naive_history.history
naive_acc = naive_dict['val_acc'] 
naive_loss = naive_dict['val_loss']

n_epochs = np.arange(1, 6)
plt.figure(figsize=(8, 6))
plt.plot(n_epochs, naive_acc, label='Validation accuracy')
plt.plot(n_epochs, naive_loss, label='Validation loss')
plt.ylabel('Validation set accuracy and loss')
plt.xlabel('Number of epochs')
plt.title('Simple sequential model validation accuracy and loss rates')
plt.legend(loc='best')
plt.show()
```

## Simple RNN

```python
# No Embedding
from keras.layers import SimpleRNN
rnn = Sequential()
rnn.add(Embedding(max_words, 50, input_length=50))
rnn.add(SimpleRNN(12))
rnn.add(Dense(1, activation='sigmoid'))
rnn.summary()

callback_lst = [
    keras.callbacks.EarlyStopping(
            monitor = 'acc',
            patience = 3),
    keras.callbacks.ModelCheckpoint(
            filepath="best_model.h5",
            monitor = "val_loss",
            save_best_only = True)]

opt = optimizers.RMSprop(lr=0.001)
rnn.compile(optimizer=opt,
              loss='binary_crossentropy',
              metrics=['acc'])
result_rnn = rnn.fit(X_train, Imp_train, epochs=50, batch_size=512,
                          callbacks = callback_lst, shuffle = True,
                          validation_data=(X_val, Imp_val), class_weight={0: 3.380773819669337, 1: 1})
```

```python
rnn_dict = result_rnn.history
rnn_acc = rnn_dict['val_acc'] 
rnn_loss = rnn_dict['val_loss']

n_epochs = np.arange(1, 6)
plt.figure(figsize=(8, 6))
plt.plot(n_epochs, rnn_acc, label='Validation accuracy')
plt.plot(n_epochs, rnn_loss, label='Validation loss')
plt.ylabel('Validation set accuracy and loss')
plt.xlabel('Number of epochs')
plt.title('Simple RNN model validation accuracy and loss rates')
plt.legend(loc='best')
plt.show()
```

### LSTM using GloVe

```python
from keras.layers import LSTM

rnn_lstm = Sequential()
rnn_lstm.add(Embedding(max_words, 50, input_length=50))
rnn_lstm.add(SimpleRNN(12, return_sequences = True))
rnn_lstm.add(LSTM(12))
rnn_lstm.add(Dense(1, activation='sigmoid'))
rnn_lstm.summary()

rnn_lstm.layers[0].set_weights([EmbedMat_glove])
rnn_lstm.layers[0].trainable = False

callback_lst = [
    keras.callbacks.ReduceLROnPlateau(
            monitor='val_acc',
            factor = 0.1,
            patience = 2),
    keras.callbacks.EarlyStopping(
            monitor = 'val_acc',
            patience = 5),
    keras.callbacks.ModelCheckpoint(
            filepath="best_model.h5",
            monitor = "val_loss",
            save_best_only = True)]

opt = optimizers.RMSprop(lr=0.0001)
rnn_lstm.compile(optimizer=opt,
              loss='binary_crossentropy',
              metrics=['acc'])
result_rnn_lstm = rnn_lstm.fit(X_train, Imp_train, epochs=50, batch_size=512,
                          callbacks = callback_lst, shuffle = True,
                          validation_data=(X_val, Imp_val), class_weight={0: 3.380773819669337, 1: 1})
```

```python
lstm_dict = result_rnn_lstm.history
lstm_acc = lstm_dict['val_acc'] 
lstm_loss = lstm_dict['val_loss']

n_epochs = np.arange(1, 12)
plt.figure(figsize=(8, 6), dpi = 500)
plt.plot(n_epochs, lstm_acc, label='Validation accuracy')
plt.plot(n_epochs, lstm_loss, label='Validation loss')
plt.ylabel('Validation set accuracy and loss')
plt.xlabel('Number of epochs')
plt.title('LSTM validation accuracy and loss rates')
plt.legend(loc='best')
plt.show()
```
